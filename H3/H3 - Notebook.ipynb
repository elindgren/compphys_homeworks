```python
import numpy as np
import matplotlib.pyplot as plt
# Set plot params
plt.rc('font', size=18)          # controls default text sizes
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=18)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # fontsize of the tick labels
plt.rc('legend', fontsize=18)    # legend fontsize
```

# Task 1

```python
def psi0_sq(x):
    return 1/np.sqrt(np.pi) * np.exp(-x**2)

data = np.loadtxt("task1.dat")
hist_data = np.loadtxt("task1_hist.dat")
dtau = 0.05
t_eq = 100
idx_eq = int(t_eq/dtau)



fig, ax = plt.subplots(figsize=(8,6))

ax.plot(data[:,0], data[:,1], linewidth=2, label=r'$N$')
ax.axhline(1000, color='C1', linewidth=3, linestyle='--', label=r'$N_0$')
ax.axvline(t_eq, color='C6', linewidth=3, linestyle=':', label=r'Equilibrium reached')
ax.legend(loc='best')
ax.set_xlabel(r'$\tau$')
ax.set_ylabel(r'$N$')
ax.grid()
# ax.set_xlim(0,50)
print(f'avg(N)={np.mean(data[idx_eq:,1]):.4f}')

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(data[:,0], data[:,2], c='C2', linewidth=2, linestyle='-.', label=r'$E_T$')
ax.axhline(0.5, color='C4', linestyle='--', linewidth=3, label=r'$E_0$')
ax.axvline(t_eq, color='C6', linewidth=3, linestyle=':', label=r'Equilibrium reached')
ax.set_xlabel(r'$\tau$')
ax.set_ylabel(r'$E_T$')
ax.legend(loc='best')
ax.grid()
print(f'avg(ET)={np.mean(data[idx_eq:,2]):.4f}')
# ax.set_xlim(0,50)

avg_hist = np.mean(hist_data[idx_eq:,1:], axis=0)
xpos = np.linspace(-5,5, hist_data.shape[1]-1)
norm = np.trapz(y=avg_hist, x=xpos)

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(x=xpos, height=avg_hist/norm, label=r'$\langle x \rangle$')
ax.plot(xpos, psi0_sq(xpos), c='k', linewidth=2, linestyle='-.', label=r'$|\psi(x)|^2$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'Probability density')
ax.legend(loc='best')
ax.grid()

print(np.trapz(y=psi0_sq(xpos), x=xpos))
print(np.trapz(y=avg_hist/norm, x=xpos))
```
